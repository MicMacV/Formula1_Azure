```
%sql
SELECT TEAM_NAME,
       COUNT(1) AS RACES,
       SUM(CALCULATED_POINTS) AS PTS,
       AVG(CALCULATED_POINTS) AS AVG_PTS
FROM F1_PRESENTATION.CALCULATED_RACE_RESULTS
GROUP BY TEAM_NAME
HAVING COUNT(1) >= 100
ORDER BY AVG_PTS DESC
```

```
%sql
--LAST DECADE
SELECT TEAM_NAME,
       COUNT(1) AS RACES,
       SUM(CALCULATED_POINTS) AS PTS,
       AVG(CALCULATED_POINTS) AS AVG_PTS
FROM F1_PRESENTATION.CALCULATED_RACE_RESULTS
WHERE RACE_YEAR BETWEEN 2011 AND 2020
GROUP BY TEAM_NAME
HAVING COUNT(1) >= 100
ORDER BY AVG_PTS DESC
```
